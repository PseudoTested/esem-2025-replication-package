```python
from pathlib import Path

import pandas as pd

categories_and_subcategories = {
    "Checks": [
        "State Check",
        "Special Case Check",
        "Output Check",
        "Input Check",
        "Trivial Output Check",
    ],
    "Update": ["State Update", "Output Update"],
    "Initialization": ["State Initialization", "Output Initialization"],
    "Return": [
        "Trivial Output Return",
        "Private Method Return",
        "Default Return",
        "Output return",
    ],
    "Other": [
        "Mathematical Computation",
        "Parent Call",
        "Data Loading",
        "Defensive Programming",
        "String Processing",
        "Scheduling",
        "Iteration",
        "Throw Exception",
    ],
}

data = [
    {
        "Category": category,
        "Subcategory": subcategory,
        "Total": 0,
        "Slicer4J Gap": 0,
        "PORBS Gap": 0,
        "PseudoSweep Gap": 0,
    }
    for category, subcategories in categories_and_subcategories.items()
    for subcategory in subcategories
]

df = pd.DataFrame(data)

files = [
    "BatchProcessor",
    "IntSquareMatrix",
    "MyStringUtils",
    "Rectangle",
    "OptimalStringAlignment",
    "WellIndented",
]

for file_name in files:
    path = Path("../negotiated-agreement-data/" + file_name + ".csv")
    df_na = pd.read_csv(path)
    df_na.dropna(subset=["What is the purpose of this code?"], inplace=True)

    for sub_cat in df["Subcategory"].values:
        df_na["What is the purpose of this code?"]
        df.loc[df.Subcategory == sub_cat, "Total"] += len(
            df_na[df_na["What is the purpose of this code?"].str.contains(sub_cat)]
        )
        df.loc[df.Subcategory == sub_cat, "Slicer4J Gap"] += len(
            df_na[
                (df_na["What is the purpose of this code?"].str.contains(sub_cat))
                & (df_na["covgap_slicer4j_slice"] == True)
            ]
        )
        df.loc[df.Subcategory == sub_cat, "PORBS Gap"] += len(
            df_na[
                (df_na["What is the purpose of this code?"].str.contains(sub_cat))
                & (df_na["covgap_porbs_slice"] == True)
            ]
        )
        df.loc[df.Subcategory == sub_cat, "PseudoSweep Gap"] += len(
            df_na[
                (df_na["What is the purpose of this code?"].str.contains(sub_cat))
                & (df_na["pseudotested_ps"] == True)
            ]
        )
print(df)
df.set_index(["Category", "Subcategory"], inplace=True)
df.to_latex(
    # buf="/Users/meganmaton/Repos/Papers/cc-orbs-ps-paper/table-data/negotiated-agreement.tex",
    index=True,
    multirow=True,
    multicolumn=True,
)
```
